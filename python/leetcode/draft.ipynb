```python
class Solution:
    def removeElement(self, nums, val):
        k = 0
        for l in nums:
            if l != val:
                nums[k] = l
                k += 1
        return k

solution = Solution()
result = solution.removeElement([3,2,2,3], 3)   
print(result)  # Output: 2
```

```python
class Solution:
    def checkIfExist(self, arr):
        for i in range(len(arr)):
            for j in range(len(arr)):
                if i != j and arr[i] == 2 * arr[j]:
                    return True
        return False

'''Input: arr = [10,2,5,3]
Output: true
Explanation: For i = 0 and j = 2, arr[i] == 10 == 2 * 5 == 2 * arr[j]
Example 2:

Input: arr = [3,1,7,11]
Output: false
Explanation: There is no i and j that satisfy the conditions.'''
```

```python
s0lution = Solution()
result = s0lution.checkIfExist([-20,8,-6,-14,0,-19,14,4])
```

```python
heights = [1,1,4,2,1,3]
hm = len(heights)
excepted = sorted(heights)
print(excepted)
```

```python
class Solution:
    def heightChecker(self, heights):
        excepted = sorted(heights)
        ans = 0
        hm = len(heights)
        for n in range(0, hm):
            if excepted[n] != heights[n]:
                ans += 1
        return ans
        

sol = Solution()
mm = sol.heightChecker([1,1,4,2,1,3])
print(mm)
```

```python
def maxProfit(self, prices):
    prof = []
    for k in range (len(prices)):
        for l in range (len(prices)-1):
            if prices[k] < prices[l]:
                prof.append(l-k)
    return max(prof)

maxProfit(None, [7,1,5,3,6,4])
```

```python
class Solution:
    def rotate(self, nums, k):
        n = len(nums)
        nums = nums[n-k:] + nums[:n-k]
        return nums
```

```python
ko = [1,2,3,4,5,6,7]
Solution.rotate(None, ko, 3)
```
